```python
for i in range(5):
    for j in range(5):
        print("j: ", j, sep="", end=" ")
    print("i: ", i, "\\n", sep="")
```

```python
# 사각형으로 별 출력하기
for i  in range(5):
    for j in range(5):
        print("*", end="")# end= 줄바꿈 대체하기
    print()
```

```python
# 사각형 모양 바꾸기
for i in range(3):
    for j in range(7):
        print("*",end="")
    print()
```

```python
# 계단식으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j < i:
            print("*", end="")
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
# 산 모양 별 출력하기 (맨 밑:9, 5층)
floor = int(input("원하는 산의 높이:"))
for i in range(floor):
    for j in range(floor):
        if j<i:
            print("*", end="")
    print()

```
