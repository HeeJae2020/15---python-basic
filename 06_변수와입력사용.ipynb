```python
# 두 숫자의 합 구하기
a, b = input("숫자 2 개 입력하세요: ").split()
a = int(a)
b = int(b)
# print(a) 
# print(b)
print(a + b)
```

```python
# 입력받은 값을 콤마를 기준으로 분리하기
a, b = map(int, input("숫자 두 개를 입력하세요").split())

# a, b = map(int, ["10", "20"])

# a, b = 10, 20

print(a + b)
```
