```python
# 반복문을호 2차원 리스트의 요소를 모두 출력하기
a = [
        [10,20],
        [30,40], 
        [50,60] 
     ]

for x, y in a:
    print(x,y)

```

```python
a = [
        [10,20],
        [30,40], 
        [50,60] 
     ]

for i in a:
    for j in i:
        print(j, end=" ")
    print()
```

```python
# for 와 range 사용하기

a = [
        [10,20],
        [30,40], 
        [50,60] 
     ]

for i in range(len(a)):
    
    for j in range(len(a[i])):

        print(a[i][j], end = " ")
    print()
```

```python
# while 반복문을 한 번 사용하기
a = [
        [10,20],
        [30,40], 
        [50,60] 
     ]

i  = 0
while i < len(a):
    x, y = a[i]
    print(x,y)
    i+=1
```

```python
# while 반복문 두 번 사용하기
a = [
        [10,20],
        [30,40], 
        [50,60] 
     ]
i = 0
while i < len(a):
    j = 0
    while j < len(a[i]):
        print(a[i][j], end=" ")
        j+=1
    print()
    i+=1
```

```python
# 톱니형 리스트 만들기
a = [3, 1, 3, 2, 5]
b = []

for i in a:
    line = []
    for j in range(i):
        line.append(0)
    b.append(line)
print(b)
```

```python
a = [3, 1, 3, 2, 5]
b = [[0] * i for i in a]
print(b)
```

```python
# 2 차원 리스트의 할당과 복사 알아보기
a = [
    [10, 20],
    [30, 40]
]
b = a
b[0][0]=500
print(a)
print(b)
```

```python
a = [
    [10, 20],
    [30, 40]
]
b = a.copy()
b[0][0] = 500
print(a)
print(b)
```

```python
x = [1, 2, 3, 4]
y = x
y[0] = 10
print(x)
print(y)
```

```python
x = [1, 2, 3, 4]
y = x.copy()
y[0] = 10
print(x)
print(y)
```

```python
import copy
a = [
    [10, 20],
    [30, 40]
]
b = copy.deepcopy(a)

b[0][0] = 500
print(a)
print(b)
```
