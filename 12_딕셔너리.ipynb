```python
# dict로 딕셔너리 만들기
lux1 = dict(health=400, mane=334, melee=550, armor=10.72)
print(lux1)
print(type(lux1))
```

```python
lux2 = dict(zip(["health", "mane", "melee", "armor"],[490, 334, 550, 10.72]))
print(lux2)
print(type(lux2))
```

```python
lux3 = dict([("health",490), ("mane", 334),("melee", 550), ("armor", 10.72)])
print(lux3)
print(type(lux3))
```
